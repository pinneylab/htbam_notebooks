## Load data


```python
def aggregate_binding_data(group: pd.DataFrame):

    intensities = group['summed_button_BGsub'].to_list()
    times = group['time_s'].to_list()
    x, y = group['x'].to_list()[0], group['y'].to_list()[0]
    id = group['id'].to_list()[0]
    series_index = group['series_index'].to_list()[0]

    # sort x and y arrays by time
    df = pd.DataFrame({'time_s': times, 'intensities': intensities})
    df.sort_values(by='time_s', inplace=True)
    times, intensities = df['time_s'].to_list(), df['intensities'].to_list()

    out = {
        'x': x,
        'y': y,
        'concentration': times,
        'intensities': intensities,
        'id': id,
        'series_index': series_index,
    }

    return pd.Series(out)
```

```python
bait_file = '/Users/jonathanzhang/Downloads/Bait_analysis.csv.bz2' # path to csv file for the bait signal generated in processing
bait_df = pd.read_csv(bait_file)
postwash_mask = bait_df['series_index'] == 'post_wash_bait' # only care about postwash bait intensity here
bait_df = bait_df[postwash_mask].groupby(by=['x', 'y']).apply(aggregate_binding_data).reset_index(drop=True)

prey_file = '/Users/jonathanzhang/Downloads/post_wash_prey.csv.bz2'
prey_df = pd.read_csv(prey_file)
prey_df = prey_df.groupby(by=['x', 'y']).apply(aggregate_binding_data).reset_index(drop=True)
```

```python
bait_intensities = np.vstack(bait_df['intensities'].to_list())
prey_intensities = np.vstack(prey_df['intensities'].to_list())
r = prey_intensities / bait_intensities

concentrations = np.vstack(prey_df['intensities'].to_list())

titles = ['({}, {}) / {}'.format(x,y,id) for x,y,id in zip(prey_df['x'], prey_df['y'], prey_df['id'])]
launch_interactive_plot(concentrations, r, titles=titles, n_rows=2, n_cols=4)
```
